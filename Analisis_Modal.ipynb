# Funciones

## Parametros Sismicos

```python
def Peru_Response_Spectrum(Tn, Soil, Zona,U,R):
    '''
    Parameters
    ----------
    Soil : String
        S0 / S1 / S2 / S3
    '''
    if   Zona == 4: Z = 0.45
    elif Zona == 3: Z = 0.35
    elif Zona == 2: Z = 0.25
    else: Z = 0.10
    
    if   Soil == "S1": S, Tp, TL = 1, 0.4, 2.5
    elif Soil == "S0": S, Tp, TL = 0.8, 0.3, 3.0
    elif Soil == "S2" and Z==0.45: S, Tp, TL = 1.05, 0.6, 2.0
    elif Soil == "S3" and Z==0.45: S, Tp, TL = 1.10, 1.0, 1.6
    elif Soil == "S2" and Z==0.35: S, Tp, TL = 1.15, 0.6, 2.0
    elif Soil == "S3" and Z==0.35: S, Tp, TL = 1.20, 1.0, 1.6
    elif Soil == "S2" and Z==0.25: S, Tp, TL = 1.20, 0.6, 2.0
    elif Soil == "S3" and Z==0.25: S, Tp, TL = 1.40, 1.0, 1.6
    elif Soil == "S2" and Z==0.10: S, Tp, TL = 1.60, 0.6, 2.0
    else: S = 2.00, 1.0, 1.6
    
    if   Tn<0.2*Tp: C = 1+7.5*(Tn/Tp)
    elif Tn<Tp:     C = 2.5
    elif Tn<TL:     C = 2.5*(Tp/Tn)
    else:           C = 2.5*((Tp*TL)/Tn**2)
    
    Sa = Z*U*S*C*981/R
    
    return Sa  #(cm/s2)
```

## Combinacion CQC

```python
def CQC(omega, response, betha = 0.05):
    '''
    Complete Quadratic Combination
    '''
    r = []
    for i in range(n_stories):
        for j in range(n_stories):
            fq_ratio = omega[j]/omega[i]
            rho=(8*betha**2*(1+fq_ratio)*fq_ratio**1.5)/((1-fq_ratio**2)**2+4*betha**2*fq_ratio*(1+fq_ratio)**2)
            r.append(response[i]*rho*response[j])
    r=np.sqrt(sum(r))
            
    return r
```

```python
"####################### Definitions ###############################"
# codigo optimizado con array-numpy
def Modal_Analysis(n,k,m):
    K = np.zeros((n,n))
    M = np.zeros((n,n))
    for i in range(n):
        if i < n-1:
            K[i,i],K[i,i+1],K[i+1,i] = k[i]+k[i+1],-k[i+1],-k[i+1]
            M[i,i] = m[i]
        else:
            K[i,i], M[i,i] = k[i], m[i]
    
    eigvals, eigvecs = la.eigh(K,M)
    omega = np.sqrt(eigvals)

    #Vibration Modes
    eigvecs = np.transpose(eigvecs)
    T =2*np.pi/omega
    #Participation Factor and Effective modal mass Ratio
    PF = eigvecs@M@np.ones(n)
    EMMR = PF**2/sum(sum(M))

    return K, M, T, omega, eigvecs, EMMR, PF
```

# Analisis Modal Espectral

```python
"####################### Structural Analysis ###############################"
Cd = 0.85*R
K, M, T, omega, modes, EMMR, PF = Modal_Analysis(n_stories,k,m)
for i in range(n_stories):
    Sai.append(Peru_Response_Spectrum(T[i],Soil,Zona,U,R))
    Svi.append(Sai[i]/(2*np.pi/T[i]))
    Sdi.append(Sai[i]/(2*np.pi/T[i])**2)

for i in range(n_stories):
    Fi.append(Sai[i]*modes[i]*PF[i]*m[i])
    Vb.append(sum(Sai[i]*modes[i]*PF[i]*m[i]))
    di.append(Sdi[i]*modes[i]*PF[i])
Fi, di = array(Fi), array(di)*(Cd)

for i in range(n_stories):
    prov, prov2 = [], [di[i][0]]
    for j in range(n_stories):
        prov.append(sum(Fi[i][j:]))
        if j < (n_stories-1):
            prov2.append(di[i][j+1]-di[i][j])
    Vi.append(prov), dri.append(prov2)
Vi, dri = array(Vi), array(dri)*(Cd/h)

V_r  = CQC(omega, Vi,  betha = 0.05)
D_r  = CQC(omega, di,  betha = 0.05)
SD_r = CQC(omega, dri, betha = 0.05)
```
